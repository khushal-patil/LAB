```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/media/khushal/New Volume/GITHUB/LAB/6TH SEMESTER/DSDBA/Assign1/data.csv")
```

```python
print(df)
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['Volume'].fillna(df['Volume'].mean(),inplace=True)
df['Weight'].fillna(df['Weight'].mean(),inplace=True)
df['CO2'].fillna(df['CO2'].mean(),inplace=True)
```

```python
print("\nStatistical Summary:")
print(df.describe())
```

```python
dimensions = df.shape
print(f"\nDataset Dimensions: {dimensions}")
```

```python
print(df.info())

# Display data types
print(df.dtypes)

print("\nConverting 'Car' and 'Model' to Categorical Variables...")
df['Car'] = df['Car'].astype('category')
df['Model'] = df['Model'].astype('category')
print(df.dtypes)

# One-hot encoding for categorical variables
print("\nApplying One-Hot Encoding to Categorical Variables...")
df_encoded = pd.get_dummies(df, columns=['Car', 'Model'], drop_first=True)
print(df_encoded.head())
```
